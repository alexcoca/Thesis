```python
import matplotlib
import matplotlib.pyplot as plt
import pickle
plt.rc('text', usetex = True)
plt.rc('font', family = 'serif')
matplotlib.rc('xtick', labelsize=14) 
matplotlib.rc('ytick', labelsize=14) 
```

Compare data obtained in the following experiments:
 
 - The density of the feature and the targets lattices is increased simulaneously from $3$ to $20$
 - The density of the targets lattice is kept constant at $k_t = 3$ and the feature lattice density is increased from $3$ to $20$
 - The density of the features lattice is kept constant at $k_f = 3$ and the targets lattice density is increased from $3$ to $20$

```python
# Load data
directories = ['D:/Thesis/Experiments/exp_target_lattice_density_feature_3/', \
               'D:/Thesis/Experiments/exp_feature_lattice_density_target_3/', \
               'D:/Thesis/Experiments/exp_2/']
filenames = ['exp_target_data', 'exp_feature_data','exp_target_feature_data' ]
data = []
for directory, filename in zip(directories, filenames):
    with open (directory + filename, 'rb') as container:
        data.append(pickle.load(container))
```

# Plot the comparisons

```python
xcoordinate = list(range(3,21))
```

## Compare optimal utilities

```python
epsilon = 0.1 # Optimal utility indep of epsilon, just a peculiarity of the storage data structure
# Independent of epsilon, all the experiments return the same utility
for experiment in data:
    plt.plot(xcoordinate, experiment['max_utilities'][epsilon], '--*')
    plt.xticks(xcoordinate)
plt.xlabel('Lattice density (points)', fontsize = 16)
plt.ylabel('max $u$', fontsize = 16)
plt.grid(True)
plt.legend([r'$k_t$ varies', r'$k_f$ varies','Both vary'],fontsize = 15)
plt.show()
```

## Compare difference between the optimal utility and the average sampled utility

```python
def delta_opt_avg_util(xcoordinate, data, epsilon):
    for experiment in data:
        plt.plot(xcoordinate, experiment['delta_opt_avg'][epsilon], '--*')
    plt.xticks(xcoordinate)
    # plt.yticks(np.arange(yticks_low, 0.02, step = 0.05))
    plt.xlabel('Target lattice denisity (points)', fontsize = 20)
    plt.ylabel(r'$opt(u) - \mu_u$ ($\varepsilon =' + str(epsilon) + '$)', fontsize = 20)
    plt.legend([r'$k_t$ varies', r'$k_f$ varies','Both vary'], fontsize = 15)
    plt.grid(True)
    plt.show()
    
eps_range = [0.1, 0.6, 5.0]
for epsilon in eps_range:
        delta_opt_avg_util(xcoordinate, data, epsilon)
```

## Compare difference between optimal utility and maximum sampled utility

```python
def delta_opt_best_util(xcoordinate, data, epsilon):
    for experiment in data:
        plt.plot(xcoordinate, experiment['delta_opt_best'][epsilon], '--*')
    plt.xticks(xcoordinate)
    # plt.yticks(np.arange(yticks_low, 0.02, step = 0.05))
    plt.xlabel('Lattice denisity (points)', fontsize = 20)
    plt.ylabel(r'$opt(u) - \max u$ ($\varepsilon =' + str(epsilon) + '$)', fontsize = 20)
    plt.legend([r'$k_t$ varies', r'$k_f$ varies','Both vary'], fontsize = 15)
    plt.grid(True)
    plt.show()
    
eps_range = [0.1, 0.6, 5.0]
for epsilon in eps_range:
        delta_opt_best_util(xcoordinate, data, epsilon)
```
